```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# ------- Lê os dados do acelerômetro -------

arquivos_corrida = [] 
arquivos_queda_livre = []

for i in range(5):
    arquivos_corrida.append(pd.read_csv(f"dados_corrida{i + 1}.csv", sep = ";"))
    arquivos_queda_livre.append(pd.read_csv(f"dados_queda{i + 1}.csv", sep = ";"))

# ------- Caminhada -------

print("\n\n================== Caminhada (MRU) ==================\n\n")

acelerometros_caminhada = []
tempos_caminhada = []
aceleracao_caminhada = []

# Marca os tempos iniciais e finais da caminhada
t0_caminhada = [3000, 4000, 4150, 2450, 11000]
tf_caminhada = [4300, 5250, 5530, 3950, 12550]

fig, ax = plt.subplots()

for i in range(5):
  acelerometros_caminhada.append(np.array(arquivos_corrida[i]))
  tempos_caminhada.append([t - acelerometros_caminhada[i][t0_caminhada[i]][0] for t in acelerometros_caminhada[i][t0_caminhada[i]:tf_caminhada[i], 0]])
  aceleracao_caminhada.append(acelerometros_caminhada[i][t0_caminhada[i]:tf_caminhada[i], 4])
  plt.plot(tempos_caminhada[i], aceleracao_caminhada[i])

plt.title("Dados do acelerômetro sobrepostos - MRU")
ax.set_xlabel("Tempo (s)")
ax.set_ylabel("Resultante (força g)")

plt.show()
print()

# ------- Corrida -------

print("\n\n================== Corrida (MUV) ==================\n\n")

acelerometros_corrida = []
tempos_corrida = []
aceleracao_corrida = []

# Marca os tempos inciais e finais da corrida
t0_corrida = [4300, 5250, 5530, 3950, 12550]
tf_corrida = [5150, 6050, 6450, 4800, 13350]

fig, ax = plt.subplots()

for i in range(5):
  acelerometros_corrida.append(np.array(arquivos_corrida[i]))
  tempos_corrida.append([t - acelerometros_corrida[i][t0_corrida[i]][0] for t in acelerometros_corrida[i][t0_corrida[i]:tf_corrida[i], 0]])
  aceleracao_corrida.append(acelerometros_corrida[i][t0_corrida[i]:tf_corrida[i], 4])
  plt.plot(tempos_corrida[i], aceleracao_corrida[i])

plt.title("Dados do acelerômetro sobrepostos - MUV")
ax.set_xlabel("Tempo (s)")
ax.set_ylabel("Resultante (força g)")

plt.show()
print()

# ------- Queda Livre -------

print("\n\n================== Queda Livre ==================\n\n")

acelerometros_queda_livre = []
tempos_queda_livre = []
aceleracao_queda_livre = []

# Marca os tempos finais e iniciais da queda livre
t0_queda = [3290, 2225, 2285, 2820, 1930]
tf_queda = [3385, 2320, 2380, 2910, 2020]

fig, ax = plt.subplots()

for i in range(5):
  acelerometros_queda_livre.append(np.array(arquivos_queda_livre[i]))
  tempos_queda_livre.append([t - acelerometros_queda_livre[i][t0_queda[i]][0] for t in acelerometros_queda_livre[i][t0_queda[i]:tf_queda[i], 0]])
  aceleracao_queda_livre.append(acelerometros_queda_livre[i][t0_queda[i]:tf_queda[i], 4])
  plt.plot(tempos_queda_livre[i], aceleracao_queda_livre[i])

plt.title("Dados do acelerômetro sobrepostos - Queda Livre")
ax.set_xlabel("Tempo (s)")
ax.set_ylabel("Resultante (força g)")

plt.show()
print()
```
